# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
from datetime import datetime
```

```python
# Read Dataset
read_data=pd.read_csv("LHR-Dubai(july 20-july 21).csv")
```

# Add Columns and Change Data Type

```python
read_data[["Departure_Time","Arrive_Time"]]=read_data['Arrival_Time'].str.split(' - ',expand=True)
```

```python
read_data["Stay"]=read_data['Journey_Duration'].str.extract(r'\((.*?)\)')
read_data['Duration']=read_data["Journey_Duration"].str.extract(r'^(\d+h(?: \d+m)?)')
```

```python
read_data['stay_time'] = read_data["Stop_Time"].str.extractall(r'(\d+h \d+m|\d+h)').groupby(level=0).agg(', '.join)
```

```python
read_data['stay_destination']=read_data["Stop_Time"].str.extractall(r'in\s(.*?)(?:\s\(|$)').groupby(level=0).agg(', '.join)
```

```python
read_data.drop(['Arrival_Time', 'Journey_Duration', 'Stop_Time'], axis=1, inplace=True)
```

```python
read_data.rename(columns={'stay_destination': 'Stay_location','stay_time':'Stay_Duration','Duration':'Total_Duration','Price':"Price(Dollars)"}, inplace=True)
```

```python
read_data['Price(Dollars)']=read_data['Price(Dollars)'].str.replace('$','')
read_data['Price(Dollars)']=read_data['Price(Dollars)'].str.replace(',','')
```

```python
# Changes Datatypes
read_data['Price(Dollars)']=read_data['Price(Dollars)'].astype(float)
read_data['Stay_Duration']=pd.to_timedelta(read_data['Stay_Duration'])
read_data['Total_Duration'] = pd.to_timedelta(read_data['Total_Duration'])
read_data['Departure_Time']=read_data['Departure_Time'].apply(lambda x: datetime.strptime(x, '%I:%M%p').strftime('%H:%M:%S'))
read_data['Departure_Time']=pd.to_timedelta(read_data['Departure_Time'])
read_data['Arrive_Time']=read_data['Arrive_Time'].apply(lambda x: datetime.strptime(x, '%I:%M%p').strftime('%H:%M:%S'))
read_data['Arrive_Time']=pd.to_timedelta(read_data['Arrive_Time'])
```

```python
exchange_rate=168.50
read_data['Price(PKR)']=read_data['Price(Dollars)']*exchange_rate
```

How does the average price vary across different airlines?

Which is the Longest Total Duration based on Airline?

```python
total_duration = read_data.groupby('AirLine_Info')['Total_Duration'].sum()
airline_with_longest_duration = total_duration.idxmax()
print("Airline with Longest Total Duration:", airline_with_longest_duration)
```

Find Outliers

```python
def find_outliers():
    price_col=read_data['Price(PKR)']
    q1=price_col.quantile(0.25)
    q3=price_col.quantile(0.75)
    IQR=q3-q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR
    outliers = price_col[(price_col < lower_bound) | (price_col > upper_bound)]
# Check if any outliers exist
    if outliers.empty:
        print("No outliers or extreme values in the Price column.")
    else:
        print("Outliers or extreme values in the Price column:")
        print(outliers)
find_outliers()
```

```python
plt.boxplot(read_data['Price(PKR)'])
plt.ylabel('Price')
plt.title('Box Plot of Price')
plt.show()
```

How does the price vary based on other factors like departure time or stay duration?

```python
df_grouped = read_data.groupby(['Departure_Time', 'Stay_Duration'])['Price(PKR)'].mean().reset_index()
plt.figure(figsize=(10, 6))
for departure_time, group in df_grouped.groupby('Departure_Time'):
    plt.plot(group['Stay_Duration'], group['Price(PKR)'], label=departure_time)

plt.xlabel('Stay Duration')
plt.ylabel('Average Price (PKR)')
plt.title('Average Price Variation by Stay Duration and Departure Time')
plt.legend(title='Departure Time', loc='best')
plt.grid(True)
plt.show()
```
